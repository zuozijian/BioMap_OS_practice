```python
# importing panda library
import pandas as pd
import json
  
# readinag given csv file
# and creating dataframe
fileinput = pd.read_csv("mentha.txt",sep=";")
  
# storing this dataframe in a csv file
fileinput.to_csv('mentha.csv', index=False)
```

```python
dataframe1 = pd.DataFrame(columns = ['A蛋白ID','A基因名称','A基因ID','A对应转录ENST','A对应蛋白ENSP',
                                     'B蛋白ID','B基因名称','B基因ID','B对应转录ENST','B对应蛋白ENSP',
                                     '来源物种','细胞类型L1','细胞类型L2','作用方向','数据来源','数据来源版本',
                                     '数据来源对应疾病','数据来源对应组织','可靠水平','质控状态','相互作用对象',
                                     '相互作用类型','数据来源相互作用类型描述','相互作用权重','相互作用置信度','备注'])
```

```python
dataframe1['A蛋白ID'] = fileinput['Protein A']
dataframe1['A基因名称'] = fileinput['Gene A']
dataframe1['B蛋白ID'] = fileinput['Protein B']
dataframe1['B基因名称'] = fileinput['Gene B']
dataframe1['来源物种'] = 'Human'
dataframe1['相互作用方向'] = 'Undirected'
dataframe1['数据来源'] = 'Mentha'
dataframe1['数据来源版本'] = 'PMID'
dataframe1['指控状态'] = 'Curated'    
dataframe1['相互作用对象'] ='PPI'
dataframe1.loc['相互作用类型'] = 'literature'
```

```python
bz = dataframe1['备注']
bz_json = bz.to_json(orient='records', force_ascii=False)
bz_json = json.loads(bz_json)
dataframe1['备注'] = bz_json
```

```python
dataframe1['备注'] = fileinput['Score'].map(lambda x:{'metha_socre':x})
```
