```python
int_a['Confidence value(s)']
```

```python
list1=[]
def int_sp(x):
    try:
        if x.split(':')[1]:
            y=x.split(':')[1]
            if float(y)>=0 and float(y)<=1:
                a=":".join(["IntAct",str(y)])
                list1.append(a)
            else:
                list1.append(None)
            
    except:
        list1.append(None)
int_a['Confidence value(s)'].map(int_sp)
```

```python
#from collections import Counter
#计算每个权重的个数
#def counter(my_list):
#    return Counter(my_list)
#counter(list1)
#list1.index(None)
int_df['相互作用权重']=list1
int_df['相互作用权重']
```
